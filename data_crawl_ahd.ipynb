```python
def merge_note_to_section(my_list):
    i = 0
    while i < len(my_list):
        if my_list[i].get('section'):
            j = i + 1
            while j < len(my_list):
                if my_list[j].get('note'):
                    my_list[i]['note'] = my_list[j]['note']
                    my_list.pop(j)
                    break
                elif my_list[j].get('section'):
                    break
                j += 1
        i += 1
    return my_list

i = 0
j = 0
```
